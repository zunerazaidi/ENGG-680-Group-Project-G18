# Scatter Plot of Min and Max Temperature (2019-2023)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Convert 'Month' and 'Year' columns to numeric with Int64 type
df['Month'] = pd.to_numeric(df['Month'], errors='coerce').astype('Int64')
df['Year'] = pd.to_numeric(df['Year'], errors='coerce').astype('Int64')

# Drop rows with NaN values in specified columns
df = df.dropna(subset=['Month', 'Year', 'Min Temp (°C)', 'Max Temp (°C)'])

# Replace infinite values with NaN and drop corresponding rows
df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=['Min Temp (°C)', 'Max Temp (°C)'])

# Create a scatter plot for Min and Max Temperature over the years
fig, ax = plt.subplots(figsize=(10, 6))
for year in range(2019, 2024):
    # Filter data for the specific year
    year_data = df[df['Year'] == year]

    # Scatter plot for Min Temperature
    ax.scatter(year_data['Month'], year_data['Min Temp (°C)'], label=f'Min Temperature {year}', alpha=0.7)

    # Scatter plot for Max Temperature
    ax.scatter(year_data['Month'], year_data['Max Temp (°C)'], label=f'Max Temperature {year}', alpha=0.7)

# Set plot title and axis labels
ax.set_title('Min and Max Temperature Scatter Plot over 2019-2023', fontsize=13, weight='bold')
ax.set_xlabel('Month')
ax.set_ylabel('Temperature (°C)')

# Set x-axis ticks and labels
ax.set_xticks(range(1, 13))
ax.set_xticklabels([
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
])

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.7)

# Place legend outside the plot area
ax.legend(bbox_to_anchor=(1, 1.01), loc='upper left')

# Show the plot
plt.show()
```

# Linear Regression

<font color='cyan'><b>a. Linear Regression for Rainfall</b></font>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import calendar

# Select relevant columns from the dataframe
selected_columns = ['Year', 'Month', 'Total Rain (mm)']
df_selected = df[selected_columns]

# Filter data for the years 2019 to 2023
df_selected = df_selected[df_selected['Year'].between(2019, 2023)]

# Separate features (X) and target variable (y)
X = df_selected[['Year', 'Month']]
y = df_selected['Total Rain (mm)']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# Standardize the features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a Linear Regression model and fit it to the training data
lm = LinearRegression()
lm.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = lm.predict(X_test_scaled)

# Calculate Mean Squared Error (MSE) for the Linear Regression model
mse_rainfall_LR = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error (Linear Regression - Rainfall): {mse_rainfall_LR:.2f}')

# Calculate accuracy based on a threshold percentage
threshold_percentage = 10.0
threshold = (threshold_percentage / 100.0) * (df['Total Rain (mm)'].max() - df['Total Rain (mm)'].min())
binary_predictions = (abs(y_pred - y_test) <= threshold).astype(int)
accuracy_rainfall_LR = binary_predictions.mean() * 100.0
print(f'Accuracy (Linear Regression - Rainfall): {accuracy_rainfall_LR:.0f}%')

# Create a scatter plot of Actual vs Predicted Rainfall for Linear Regression
plt.figure(figsize=(11.62, 6))
plt.scatter(X_test.index, y_test, color='blue', label='Actual', alpha=0.5)
plt.scatter(X_test.index, y_pred, color='green', label='Predicted', alpha=0.5)
plt.xlabel('Index')
plt.ylabel('Rainfall')
plt.title('Actual vs Predicted (Rainfall - Linear Regression)', fontsize=13, weight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1, 1.0), loc='upper left')
plt.show()

# Combine features and target variable into a new dataframe
df_combined = pd.concat([X, y], axis=1)

# Convert 'Year' and 'Month' columns to appropriate data types
df_combined['Year'] = df_combined['Year'].astype(int)
df_combined['Month'] = df_combined['Month'].astype(int)

# Convert numeric month values to abbreviated month names
df_combined['Month'] = df_combined['Month'].apply(lambda x: calendar.month_abbr[x])

# Create a scatter plot with Year-wise coloring for Rainfall
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(x='Month', y='Total Rain (mm)', hue='Year', data=df_combined, palette='bright', hue_order=df_combined['Year'].unique())
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.title('Scatter Plot of Rain (mm) against Month with Year-wise Coloring (Linear Regression)', fontsize=13, weight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1, 1.0), loc='upper left')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import calendar

# Selecting relevant columns
selected_columns = ['Year', 'Month', 'Avg Temp (°C)']
df_selected = df[selected_columns]

# Filtering data for the years 2019 to 2023
df_selected = df_selected[df_selected['Year'].between(2019, 2023)]

# Extracting features (X) and target variable (y)
X = df_selected[['Year', 'Month']]
y = df_selected['Avg Temp (°C)']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# Polynomial regression with degree 2
degree = 2
polyreg = make_pipeline(PolynomialFeatures(degree), StandardScaler(), LinearRegression())
polyreg.fit(X_train, y_train)

# Predicting on the test set
y_pred_poly = polyreg.predict(X_test)

# Calculating Mean Squared Error (MSE)
mse_avg_temp_LR = mean_squared_error(y_test, y_pred_poly)
print(f'Mean Squared Error (Linear Regression - Average Temperature): {mse_avg_temp_LR:.2f}')

# Calculating Accuracy using a threshold
threshold_percentage_poly = 5.0
threshold_poly = (threshold_percentage_poly / 100.0) * (df['Avg Temp (°C)'].max() - df['Avg Temp (°C)'].min())
binary_predictions_poly = (abs(y_pred_poly - y_test) <= threshold_poly).astype(int)
accuracy_avg_temp_LR = binary_predictions_poly.mean() * 100.0
print(f'Accuracy (Linear Regression - Average Temperature): {accuracy_avg_temp_LR:.0f}%')

# Creating a scatter plot of Actual vs Predicted values
plt.figure(figsize=(11.62, 6))
plt.scatter(X_test.index, y_test, color='blue', label='Actual', alpha=0.5)
plt.scatter(X_test.index, y_pred, color='green', label='Predicted', alpha=0.5)
plt.xlabel('Index')
plt.ylabel('Average Temperature')
plt.title('Actual vs Predicted (Average Temperature - Linear Regression)', fontsize=13, weight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1, 1.0), loc='upper left')
plt.show()

# Combining features and target variable for visualization
df_combined = pd.concat([X, y], axis=1)

# Converting 'Year' and 'Month' to integers and abbreviating 'Month' to three-letter format
df_combined['Year'] = df_combined['Year'].astype(int)
df_combined['Month'] = df_combined['Month'].astype(int)
df_combined['Month'] = df_combined['Month'].apply(lambda x: calendar.month_abbr[x])

# Creating a scatter plot with Year-wise coloring
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(x='Month', y='Avg Temp (°C)', hue='Year', data=df_combined, palette='bright', hue_order=df_combined['Year'].unique())
plt.xlabel('Month')
plt.ylabel('Average Temperature')
plt.title('Scatter Plot of Average Temperature (°C) against Month with Year-wise Coloring (Linear Regression)', fontsize=13, weight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1, 1.0), loc='upper left')
plt.show()
```

#  K-Nearest Neighbors (KNN)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import calendar
from matplotlib.cm import ScalarMappable

selected_columns = ['Year', 'Month', 'Total Rain (mm)']
df_selected = df[selected_columns]

df_selected = df_selected[df_selected['Year'].between(2019, 2023)]

X = df_selected[['Year', 'Month']]
y = df_selected['Total Rain (mm)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

k = 5
knn_regressor = KNeighborsRegressor(n_neighbors=k)
knn_regressor.fit(X_train_scaled, y_train)

y_pred = knn_regressor.predict(X_test_scaled)

mse_rainfall_KNN = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error (KNN - Rainfall): {mse_rainfall_KNN:.2f}')

threshold_percentage = 10.0
threshold = (threshold_percentage / 100.0) * (df['Total Rain (mm)'].max() - df['Total Rain (mm)'].min())
binary_predictions = (abs(y_pred - y_test) <= threshold).astype(int)
accuracy_rainfall_KNN = binary_predictions.mean() * 100.0
print(f'Accuracy (KNN - Rainfall): {accuracy_rainfall_KNN:.0f}%')

plt.figure(figsize=(11.62, 6))
plt.scatter(X_test.index, y_test, color='blue', label='Actual', alpha=0.5)
plt.scatter(X_test.index, y_pred, color='green', label='Predicted', alpha=0.5)
plt.xlabel('Data Point')
plt.ylabel('Rainfall')
plt.title('Actual vs Predicted (Rainfall - KNN Regression)', fontsize=13, weight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1, 1.0), loc='upper left')
plt.show()

df_combined = pd.concat([X, y], axis=1)

df_combined['Year'] = df_combined['Year'].astype(int)
df_combined['Month'] = df_combined['Month'].astype(int)

df_combined['Month'] = df_combined['Month'].apply(lambda x: calendar.month_abbr[x])

plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(x='Month', y='Total Rain (mm)', hue='Year', data=df_combined, palette='bright', hue_order=df_combined['Year'].unique())
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.title('Scatter Plot of Rain (mm) against Month with Year-wise Coloring (KNN)', fontsize=13, weight='bold')

plt.grid(True, linestyle='--', alpha=0.7)

plt.legend(bbox_to_anchor=(1, 1.0), loc='upper left')

plt.show()
```

```python
import matplotlib.pyplot as plt

algorithms = ['Linear Regression', 'KNN', 'Random Forest', 'Gradient Boosting']

# Mean Squared Error for each algorithm and parameter
mse_rainfall = [mse_rainfall_LR, mse_rainfall_KNN, mse_rainfall_RF, mse_rainfall_GB]
mse_snowfall = [mse_snowfall_LR, mse_snowfall_KNN, mse_snowfall_RF, mse_snowfall_GB]
mse_precip = [mse_precip_LR, mse_precip_KNN, mse_precip_RF, mse_precip_GB]
mse_avg_temp = [mse_avg_temp_LR, mse_avg_temp_KNN, mse_avg_temp_RF, mse_avg_temp_GB]

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bar width
bar_width = 0.2

# Set position for each group of bars
index = range(len(algorithms))
index_rainfall = [i - bar_width for i in index]
index_snowfall = index
index_precip = [i + bar_width for i in index]
index_avg_temp = [i + 2 * bar_width for i in index]

# Plot bars for each parameter
plt.bar(index_rainfall, mse_rainfall, width=bar_width, label='Rainfall')
plt.bar(index_snowfall, mse_snowfall, width=bar_width, label='Snowfall')
plt.bar(index_precip, mse_precip, width=bar_width, label='Precipitation')
plt.bar(index_avg_temp, mse_avg_temp, width=bar_width, label='Average Temperature')

# Set labels and title
plt.xlabel('Algorithms')
plt.ylabel('Mean Squared Error')
plt.title('Mean Squared Error for Prediction')
plt.xticks(index, algorithms)
plt.legend(bbox_to_anchor=(1, 1.0), loc='upper left')

# Show the plot
plt.show()
```

<font color='orange'><b>Explanation of the Code</b></font>

This code creates a grouped bar plot to compare the Mean Squared Error (MSE) of different machine learning algorithms (Linear Regression, KNN, Random Forest, and Gradient Boosting) for predicting four weather parameters: Rainfall, Snowfall, Precipitation, and Average Temperature. The MSE values for each algorithm and parameter are stored in separate lists. Using Matplotlib, the script sets up the figure and axis, defines the bar width, and specifies the position for each group of bars. It then plots the MSE values for each parameter, arranging them side by side for better comparison. The x-axis represents the algorithms, and the y-axis represents the MSE. The resulting plot provides a visual comparison of the prediction performance across algorithms and weather parameters.
